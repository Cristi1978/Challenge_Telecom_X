```python
# Normalizar las columnas
from pandas import json_normalize

df_telecomx_normalized = json_normalize(data)
```

```python
columnas_norm = ['customer', 'phone', 'internet', 'account']

df_telecomx_normalized = pd.DataFrame()
for columna in columnas_norm:
 df_telecomx_normalized = pd.concat([df_telecomx_normalized, pd.json_normalize(df_telecomx[columna])], axis=1)
df_telecomx_normalized = pd.concat([df_telecomx.drop(columns=columnas_norm),df_telecomx_normalized ], axis=1)
df_telecomx_normalized
```

```python
df_telecomx_normalized.rename(columns={
    'customerID': 'id',
    'Churn':'cancelacion',
    'gender': 'genero',
    'SeniorCitizen': 'mayor_de_65',
    'Partner': 'pareja',
    'Dependents': 'dependientes',
    'tenure': 'meses_de_contrato',
    'PhoneService': 'servicio_telefonico',
    'MultipleLines': 'lineas_multiples',
    'InternetService': 'servicio_internet',
    'OnlineSecurity': 'seguridad_en_linea',
    'OnlineBackup': 'Soporte_en_linea',
    'DeviceProtection': 'protecc_disp',
    'TechSupport': 'soporte_tecnico',
    'StreamingTV': 'servicio_tv',
    'StreamingMovies': 'servicio_peliculas',
    'PaperlessBilling': 'facturas_electronicas',
    'PaymentMethod': 'metodo_pago',
    'Contract': 'tipo_contrato',
    'Charges.Monthly': 'total_pagos_mensuales',
    'Charges.Total': 'ingreso_total_cliente'
}, inplace=True)
```

 ### Visualizaci√≥n que muestra la Cancelaci√≥n


```python
ax = sns.countplot(x='cancelacion', data=df_telecomx_normalized, hue='cancelacion', palette=['#2ca02c', '#ff7f0e'], legend=False)
plt.title('Proporci√≥n de clientes que permanecen vs los que abandonaron')
plt.ylabel('Cantidad')
plt.xticks([0, 1], ['Permanecen', 'Abandonan'])

total = len(df_telecomx_normalized)
for p in ax.patches:
    altura = p.get_height()
    ax.annotate(f'{altura}\n({altura/total:.1%})',
                (p.get_x() + p.get_width() / 2., altura * 0.95), # Adjust vertical position
                ha='center', va='top', fontsize=10, fontweight='bold') # Align text to the top
plt.show()
```

```python
colores = ['#2ca02c', '#ff7f0e']  # Permanecen, Abandonan

variables_categoricas = [
    'genero', 'pareja', 'dependientes', 'servicio_telefonico',
    'lineas_multiples', 'servicio_internet', 'seguridad_en_linea',
    'Soporte_en_linea', 'protecc_disp', 'soporte_tecnico',
    'servicio_tv', 'servicio_peliculas', 'tipo_contrato',
    'facturas_electronicas', 'metodo_pago'
]

for i in range(0, len(variables_categoricas), 4):
    fig, axs = plt.subplots(2, 2, figsize=(14, 10))
    axs = axs.flatten()

    for j in range(4):
        if i + j < len(variables_categoricas):
            col = variables_categoricas[i + j]
            ax = axs[j]

            # Obtenemos las cantidades absolutas y proporciones
            conteos = df_telecomx_normalized.groupby(col)['cancelacion'].value_counts().unstack().fillna(0)
            proporciones = conteos.div(conteos.sum(axis=1), axis=0)

            # Graficar proporciones
            proporciones.plot(kind='bar', stacked=True, color=colores, ax=ax)

            # Agregar etiquetas de cantidad en cada barra
            for idx, cat in enumerate(conteos.index):
                total = conteos.loc[cat]
                bottom = 0
                for k, val in enumerate(total):
                    height = proporciones.loc[cat][k]
                    ax.text(
                        idx, bottom + height / 2,
                        f'{int(val)}',
                        ha='center', va='center', color='white', fontsize=9, fontweight='bold'
                    )
                    bottom += height

            ax.set_title(f'Proporci√≥n de Abandono por {col.replace("_", " ").title()}')
            ax.set_ylabel('Proporci√≥n')
            ax.set_xlabel(col.replace("_", " ").title())

            ax.legend(['Permanecen', 'Abandonan'])
        else:
            fig.delaxes(axs[j])

    plt.tight_layout()
    plt.show()
```

#**Conclusiones y Recomendaciones**

En resumen, este an√°lisis ha identificado varios factores clave que est√°n fuertemente asociados con la cancelaci√≥n de clientes en la empresa de telecomunicaciones:

‚Ä¢	**Tipo de Contrato**: Los contratos mes a mes tienen la tasa de cancelaci√≥n m√°s alta.

‚Ä¢	**M√©todo de Pago**: El pago electr√≥nico se asocia con una mayor probabilidad de cancelaci√≥n.

‚Ä¢	**Servicio de Internet**: Los clientes con servicio de fibra √≥ptica muestran una tasa de cancelaci√≥n m√°s alta.

‚Ä¢	**Demograf√≠a**: Los adultos mayores, los clientes sin pareja y sin dependientes tienden a cancelar m√°s.

‚Ä¢	**Servicios Adicionales**: La falta de servicios adicionales, especialmente aquellos relacionados con internet, est√° vinculada a una mayor tasa de cancelaci√≥n.

Basado en estos hallazgos, se sugieren las siguientes recomendaciones para reducir la tasa de cancelaci√≥n:

‚Ä¢	**Incentivar Contratos a Largo Plazo**: Ofrecer descuentos o beneficios exclusivos a los clientes que elijan contratos de uno o dos a√±os para fomentar la lealtad.

‚Ä¢	**Optimizar la Experiencia de Pago Electr√≥nico**: Investigar las posibles razones detr√°s de la alta tasa de cancelaci√≥n asociada al pago electr√≥nico y mejorar la experiencia del usuario en esta plataforma.

‚Ä¢	**Mejorar la Calidad del Servicio de Fibra √ìptica**: Abordar cualquier problema t√©cnico o de rendimiento percibido por los clientes de fibra √≥ptica para reducir su insatisfacci√≥n.

‚Ä¢	**Programas de Retenci√≥n Dirigido**: Desarrollar programas de retenci√≥n espec√≠ficos para segmentos de clientes con mayor riesgo de cancelaci√≥n, como adultos mayores, clientes individuales y aquellos sin servicios adicionales, ofreciendo atenci√≥n personalizada o promociones relevantes.

‚Ä¢	**Promocionar Servicios Adicionales**: Resaltar los beneficios de los servicios adicionales, especialmente los de seguridad y soporte, para aumentar su adopci√≥n y potencialmente mejorar la retenci√≥n, particularmente entre los usuarios de internet.

Implementar estas recomendaciones de manera estrat√©gica podr√≠a ayudar a la empresa a reducir la p√©rdida de clientes y mejorar la satisfacci√≥n general.
